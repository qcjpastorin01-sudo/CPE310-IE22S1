**Exercise 3:** Rewrite your pay computation with time-and-a-half for overtime and create a function called computepay which takes two parameters (hours and rate).



```
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```



````python
def computepay(hours, rate):
    if hours <= 40:
        return hours * rate
    else:
        overtime = hours - 40
        return 40 * rate + overtime * rate * 1.5

hours = float(input("Enter hours: "))
rate = float(input("Enter rate: "))

print("Pay:", computepay(hours, rate))
````

## **Part 2: Loops and Iterations**

**Exercise 1:** Write a program which repeatedly reads integers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the integers. If the user enters anything other than a integers, detect their mistake using try and except and print an error message and skip to the next integers.

````python
totalcount = 0
sum = 0
average = 0

count = input("Enter a number (or 'done' to finish): ")
while count != "done":
  try:
    integernum = int(count)
    sum += integernum
    totalcount += 1
    average = sum / totalcount
  except:
    print("Invalid input. Please enter an integer.")

  count = input("Enter a number (or 'done' to finish): ")
print("________________________________________________")
print("Total inputs: ", totalcount)
print("Total sum: ", sum)
print("Average: ", average)
````

**Exercise 2:** Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

````python
totalcount = 0
sum = 0
max_num = None
min_num = None

count = input("Enter a number (or 'done' to finish): ")
while count != "done":
  try:
    integernum = int(count)
    sum += integernum
    totalcount += 1

    if max_num is None or integernum > max_num:
      max_num = integernum
    if min_num is None or integernum < min_num:
      min_num = integernum

  except:
    print("Invalid input. Please enter an integer.")

  count = input("Enter a number (or 'done' to finish): ")

print("________________________________________________")
print("Total inputs:", totalcount)
print("Total sum:", sum)
print("Maximum:", max_num)
print("Minimum:", min_num)
````
