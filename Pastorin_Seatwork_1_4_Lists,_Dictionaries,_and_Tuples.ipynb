**Python Lists: Built-In Data** <br>

List - is a built-in data structure

image.png

```python
#1. Create a list
fruits = ["apple", "banana", "cherry", "orange"]
print("Original list: ", fruits)
```

```python
#2. Create Elements
print("First fruit: ", fruits[0])
print("Last fruit: ", fruits[-1])
```

```python
#3. Add Elements
fruits = ["apple", "banana", "cherry", "orange"]
fruits.append("mango")     # Add at the end
fruits.insert(1, "grape")  # Add at specific position
print("After adding fruits:", fruits)
```

```python
#4. Remove Elements
fruits.remove("banana")                          # Remove by value
popped = fruits.pop()
print("After removing fruits:", fruits)          # Remove last item
print("Popped item:", popped)
```

```python
#5. Check if an item exists
if "apple" in fruits:
  print("Apple is in the list")
```

```python
#6. Loop through the list
print("All fruits in the list!")
for fruit in fruits:
  print("-", fruit)
```

```python
#7. Length of the list
print("Total number of fruits:", len(fruits))
```

```python
#8. Sorting the list
fruits.sort()
print("Sorted list:", fruits)
```

```python
#9. Reversing the list
fruits.reverse()         # Alphabetical
print("Reversed list", fruits)
```

**DYNAMIC MAPPING: DICTIONARIES** <br>
Dictionaries and mutable, indexed collections consisting of key: value pairs

```python
pol_eng_dictionary = {
    "kwiat": "flower",
    "woda": "water",
    "gleba": "soil"
  }

item_1 = pol_eng_dictionary["gleba"]          # ex. 1
print(item_1)                                 # outputs: soil

item_2 = pol_eng_dictionary.get("woda")      # ex. 2
print(item_2)                                 # outputs: water
```
