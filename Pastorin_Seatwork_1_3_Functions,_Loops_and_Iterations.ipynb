**WHILE LOOP** - Iterate as long as a condition remains true. (while test_expression) - 'di alam saan mah eend. <br>
**FOR LOOP** - Traverses a squence (List, String, Range). Continues until the end of the sequence. - alam kung saan mag eend. <br>

array (C++) is called ***lists*** in python. eg. [apple. banana, cherry]

```python
#1. For loop
print("For lopp:")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print("-", fruit)
```

```python
#2. While loop
print("\nWhile loop:")
count = 1
while count <= 5:
  print("Count:", count)
  count += 1
```

**LOOP CONTROL** = After loop flow dynamically.

*   **Break** - *terminates* the entire loop immediately. (end)
*   **Continue** - *skips* omly the remaining code in the current iteration and jumps to the next cycle.

```python
#Continue
for i in range(1,6):
  if i == 3:
    continue
  print(i)
```

```python
#Break
for i in range(1, 6):
  if i == 3:
    break
  print(i)
```
